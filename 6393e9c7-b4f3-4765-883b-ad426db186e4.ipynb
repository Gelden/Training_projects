````python
from scipy import stats as st
game_from_2013_xone = game_from_2013.query('platform == "XOne"')
game_from_2013_pc = game_from_2013.query('platform == "PC"')
game_from_2013_xone['user_score'] = game_from_2013_xone['user_score'].fillna(game_from_2013_xone.groupby('genre')['user_score'].transform('median')) 
#замена NaN на медиану на основе группировки по жанру
game_from_2013_pc['user_score'] = game_from_2013_pc['user_score'].fillna(game_from_2013_pc.groupby('genre')['user_score'].transform('median')) 
#замена NaN на медиану на основе группировки по жанру
game_from_2013_pc['user_score'] = game_from_2013_pc['user_score'].fillna(value=0)
#после проведенной операции остались 4 пропуска, они были заменены на 0
alpha = 0.05
results = st.ttest_ind(game_from_2013_xone['user_score'], game_from_2013_pc['user_score'], equal_var = False)

print('p-значение: ', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
import warnings
warnings.filterwarnings('ignore')
````
